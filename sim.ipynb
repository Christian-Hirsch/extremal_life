# Simulation study for the maximally persistent cycle

Start with a point cloud distributed entirely at random in a sampling window. In this notebook, we investigate the question:


**How long is the maximally persistent cycle?**

For the simulation study, we rely on a Poisson point process with intensity 2 in a $15 \times 15$ sampling window.

```r
rm(list = ls())

library(ggplot2)
library(spatstat)
library(TDA)
```

```r
seed <- 1
set.seed(seed)

width = 20
height = 10
options(repr.plot.width = width, 
        repr.plot.height = height)

#number of samples
nsim <- 1e3

#sampling window
win = owin(c(0,15), 
           c(0,15))

#poisson process 
ppps <- rpoispp(lambda = 2, 
                win = win,
                nsim = nsim)

ppp <- ppps[[1]]
qplot(x = ppp$x,  
      y = ppp$y, 
      size = I(2.5)) + 
        theme_void() +
        theme(plot.margin = margin(10, 10, 10, 10)) +
        coord_fixed() +
        theme(panel.border = element_rect(fill = NA))
```

Persist the data to csv.

```r
df <- data.frame(x = ppp$x,
                 y = ppp$y)
write.csv(df, 'ppp.csv')
```

## Persistence Diagram

Compute persistence diagram and extract locations.

```r
acf <- alphaComplexFiltration(cbind(ppp$x, ppp$y))
cd <- filtrationDiag(acf, 
                     maxdimension=1, 
                     library = "Dionysus", 
                     location = TRUE)
locs <- cd$cycleLocation
```

Get maximal cycle and minimal cycle

```r
locs
```

```r
life_time_idxs <- order(cd$diagram[,'Death'] - cd$diagram[,'Birth'])

max_idx <- rev(life_time_idxs)[2]
typ_idx <- 517
min_idx <- life_time_idxs[1]

#compute starting points and endpoints of edges
                            
max_start_idxs <- locs[[max_idx]][,1]
max_end_idxs <- locs[[max_idx]][,2]   

typ_start_idxs <- locs[[typ_idx]][,1]
typ_end_idxs <- locs[[typ_idx]][,2]  

min_start_idxs <- locs[[min_idx]][,1]
min_end_idxs <- locs[[min_idx]][,2]   
```

Define dataframes.

```r
df_max <- data.frame(start_x = ppp$x[max_start_idxs],
                 start_y = ppp$y[max_start_idxs],
                 end_x = ppp$x[max_end_idxs],
                 end_y = ppp$y[max_end_idxs])

df_typ <- data.frame(start_x = ppp$x[typ_start_idxs],
                 start_y = ppp$y[typ_start_idxs],
                 end_x = ppp$x[typ_end_idxs],
                 end_y = ppp$y[typ_end_idxs])

df_min <- data.frame(start_x = ppp$x[min_start_idxs],
                 start_y = ppp$y[min_start_idxs],
                 end_x = ppp$x[min_end_idxs],
                 end_y = ppp$y[min_end_idxs])
```

Persist to csv

```r
write.csv(df_max, 
          'max_cycle_.csv')
write.csv(df_typ, 
          'typ_cycle_.csv')
write.csv(df_min, 
          'min_cycle_.csv')
```
